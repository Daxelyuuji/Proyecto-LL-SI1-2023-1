```python
import ipywidgets as widgets
from IPython.display import display

def calcular_faltante(numero):
    if numero < 10:
        return f"El número {numero} tiene una cifra. Le falta {10 - numero} para ser un número de 2 cifras."
    elif numero < 100:
        return f"El número {numero} tiene dos cifras. Le falta {100 - numero} para ser un número de 3 cifras."
    else:
        return f"El número {numero} tiene tres cifras o más."

def on_button_clicked(b):
    try:
        numero = int(numero_text.value)
        resultado_text.value = calcular_faltante(numero)
    except ValueError:
        resultado_text.value = "Por favor, ingrese un número válido."

numero_text = widgets.Text(description="Número:")
button = widgets.Button(description="Calcular")
resultado_text = widgets.Textarea()

button.on_click(on_button_clicked)

display(numero_text, button, resultado_text)
```

```python
import ipywidgets as widgets
from IPython.display import display

def calcular_descuento(color, total_compra):
    descuento = 0
    if color == "verde":
        descuento = total_compra * 0.10
    elif color == "amarilla":
        descuento = total_compra * 0.25
    elif color == "azul":
        descuento = total_compra * 0.50
    elif color == "roja":
        descuento = total_compra
    total_pagar = total_compra - descuento
    return total_pagar
def on_button_clicked(b):

    try:
        total_compra = float(total_compra_text.value)
        color = colores_dropdown.value
        total_pagar = calcular_descuento(color, total_compra)
        resultado_text.value = f"Total a pagar con descuento: S/{total_pagar}"
    except ValueError:
        resultado_text.value = "Por favor, ingrese un valor válido para la compra."

colores = ["blanco", "verde", "amarilla", "azul", "roja"]
colores_dropdown = widgets.Dropdown(options=colores, description="Color de la bolita:")
total_compra_text = widgets.Text(description="Total de la compra:")
button = widgets.Button(description="Calcular descuento")
resultado_text = widgets.Textarea()
button.on_click(on_button_clicked)
display(colores_dropdown, total_compra_text, button, resultado_text)
```

```python
N = int(input("Ingrese el número de empleados: "))

estatura_H = []
empleadas = 0
estatura_M = []
sueldos_M = []

for i in range(N):
    genero = int(input("Ingrese el género del empleado (1: Femenino, 2: Masculino): "))
    estatura = float(input("Ingrese la estatura del empleado en centímetros: "))
    sueldo = float(input("Ingrese el sueldo del empleado: "))

    if genero == 1:  # Femenino
        empleadas += 1
        estatura_M.append(estatura)
        sueldos_M.append(sueldo)
    elif genero == 2:  # Masculino
        estatura_H.append(estatura)

estatura_prom_H = sum(estatura_H) / len(estatura_H) if estatura_H else 0
porcentaje_M = (empleadas / N) * 100 if N else 0
min_estatura_M = min(estatura_M) if estatura_M else 0
prom_sueldos_M = sum(sueldos_M) / len(sueldos_M) if sueldos_M else 0

print("a) Estatura promedio de los empleados masculinos:", estatura_prom_H, "cm")
print("b) Porcentaje de empleados femeninos que trabajan en la fábrica:", porcentaje_M, "%")
if empleadas > 0:
    print("c) La menor estatura entre el personal femenino:", min_estatura_M, "cm")
    print("d) Promedio de sueldos del personal femenino:", prom_sueldos_M)
else:
    print("No hay empleados femeninos en la fábrica.")
```

```python
# Inicializamos variables
totalP = 0
totalH = 0
totalM = 0
sumedadesH = 0
sumedadesM = 0
edmasjoven = float('inf')
# Solicitar información hasta que se responda negativamente a la pregunta de si desea agregar otra persona
while True:
    edad = int(input("Ingrese la edad de la persona: "))

    while True:
        genero = input("Ingrese el género de la persona (F para femenino, M para masculino): ").upper()
        if genero not in ['F', 'M']:
            print("Género no válido. Por favor, ingrese F o M.")
            continue
        else:
            break
    # Contar personas por género
    if genero == 'M':
        totalH += 1
        sumedadesH += edad
    elif genero == 'F':
        totalM += 1
        sumedadesM += edad
    # Actualizar la edad más joven
    if edad < edmasjoven:
        edmasjoven = edad
    totalP += 1
    # Preguntar si desea agregar otra persona
    respuesta = input("¿Desea agregar otra persona? (Sí/No): ").lower()
    if respuesta != 'sí' and respuesta != 'si':
        break
# Calcular promedio de edades por género
promedadH = sumedadesH / totalH if totalH > 0 else 0
promedadM = sumedadesM / totalM if totalM > 0 else 0
# Mostrar resultados
print("\n--- Resultados de la Conferencia ---")
print("a) Total de personas que asistieron al congreso:", totalP)
print("b) Total de hombres:", totalH)
print("   Total de mujeres:", totalM)
print("c) Promedio de edades por género:")
print("   - Promedio de edad de hombres:", promedadH)
print("   - Promedio de edad de mujeres:", promedadM)
print("d) Edad de la persona más joven que asistió al congreso:", edmasjoven)
```
